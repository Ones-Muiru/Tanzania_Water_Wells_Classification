The [total static head](https://www.pumpfundamentals.com/download/book/chapter3.pdf) is the difference between the discharge static head and the suction static head, or the difference in elevation at the outlet of a well or borehole including the pressure head at the outlet, and the elevation at the inlet including the pressure head at the inlet.

Here are the reasons why I argue that there are no outliers or missing values in amount_tsh column.

1. Availability of water in a well changes with the seasons and I don't know if some of the wells had their data recorded during a period of heavy rainfall while others had their data recorded during the dry season. 
I also don't know the units of measure that were used for this column.

2. The high number of zero values are making non-zero values appear as outliers. 

3. JICA (Japan International Cooperation Agency) has reported that [some wells in the greater Dar es Salaam region can deliver at least 100 liters of water per minute](https://openjicareport.jica.go.jp/pdf/11814811_01.pdf), which amounts to about 144,000 litres in a day. Therefore, the high values of 350,000 may not be unrealistic.

4. From the box plot of total_static_head vs. functionality labels, we can see that the pumps having tsh above approx.125,000 are all functional. High static head may be an important feature because the higher the tsh the higher the probabilty of a pump being functional.

Because of these reasons, I choose to retain all the rows as they are in this column.

```python
# Create function to print folium map
def folium_map(data, url):
    # Create 3 dataframes for each status_group
    df_f = df[df['status_group'] == 'functional']
    df_nf = df[df['status_group'] == 'non functional']
    df_fnr = df[df['status_group'] == 'functional needs repair']
    
    # Create lists of latitude and longitude values
    lat_f = [x for x in df_f['latitude']]
    long_f = [x for x in df_f['longitude']]

    lat_nf = [x for x in df_nf['latitude']]
    long_nf = [x for x in df_nf['longitude']]

    lat_fnr = [x for x in df_fnr['latitude']]
    long_fnr = [x for x in df_fnr['longitude']]

    lat_long_f = [(lat_f[i], long_f[i]) for i in range(len(lat_f))]
    lat_long_nf = [(lat_nf[i], long_nf[i]) for i in range(len(lat_nf))]
    lat_long_fnr = [(lat_fnr[i], long_fnr[i]) for i in range(len(lat_fnr))]
    
    # Create map for all locations
    this_map = folium.Map()

    # Loop through 3 dataframes and plot points for each coordinate with labels
    for coord in lat_long_nf[::5]:
        folium.CircleMarker(location=[coord[0], coord[1]], opacity=0.6, color='orange', 
                            radius=1, fill_color='orange', fill_opacity=0.6, 
                            fill=True, tooltip='non functional').add_to(this_map)
    for coord in lat_long_f[::5]:
        folium.CircleMarker(location=[coord[0], coord[1]], opacity=0.6, color='blue', 
                            radius=1, fill_color='blue', fill_opacity=0.6, 
                            fill=True, tooltip='functional').add_to(this_map)
    for coord in lat_long_fnr[::5]:
        folium.CircleMarker(location=[coord[0], coord[1]], opacity=0.6, color='green', 
                            radius=1, fill_color='green', fill_opacity=0.6, 
                            fill=True, tooltip='functional needs repair').add_to(this_map) 

    # Set the zoom to fit our bounds
    this_map.fit_bounds(this_map.get_bounds())

    # Add a custom legend
    legend_html = """
    <div style="position: fixed; 
                 bottom: 50px; left: 50px; width: 200px; height: 100px; 
                 background-color: white;
                 border:2px solid grey; z-index:9999; font-size:14px;
                 ">&nbsp; Legend <br>
                 &nbsp; <i class="fa fa-circle" style="color:orange"></i> Non Functional &nbsp; <br>
                 &nbsp; <i class="fa fa-circle" style="color:blue"></i> Functional &nbsp; <br>
                 &nbsp; <i class="fa fa-circle" style="color:green"></i> Functional Needs Repair &nbsp;
    </div>
    """

    this_map.get_root().html.add_child(folium.Element(legend_html))

    this_map.save(url)
    return this_map

folium_map(df, 'images/distribution_of_total_wells_in_tz.html')
```

#### <span style="color:blue">4.3.1 MODEL PERFORMANCE CLASS</span>	<a class="anchor" id="thirtyseventh-bullet"></a>

```python
# Create a class to print model scores
# Initiate metrics list as a global variable
metrics_list=[]

class Metrics:
    def model_score(self, model, y_true, y_pred):
        # Print classification report, accuracy, precision, recall, f1_score 
        print(classification_report(y_true, y_pred))
        print("Overall accuracy score", accuracy_score(y_true, y_pred))
        print("Overall precision score", precision_score(y_true, y_pred, average='weighted'))
        print("Overall recall score", recall_score(y_true, y_pred, average='weighted'))
        print("Overall F1-score", f1_score(y_true, y_pred, average='weighted'))

        # Print a confusion matrix    
        cnf_matrix = confusion_matrix(y_true, y_pred)
        disp = ConfusionMatrixDisplay(confusion_matrix=cnf_matrix, display_labels=model.classes_)
        print('\nConfusion Matrix')       
        return disp.plot()

    # Create a list of model metrics
    def get_metrics(self, model_name, model, y_true, y_pred):  #y_test, X_test, model
        metrics = {}  
        metrics['model_name'] = model_name
        metrics['accuracy'] = accuracy_score(y_true, y_pred)
        metrics['f1 score'] = f1_score(y_true, y_pred, average='weighted')
        metrics['precision'] = precision_score(y_true, y_pred, average='weighted')
        metrics['recall'] = recall_score(y_true, y_pred, average='weighted')
        metrics_list.append(metrics)
        return metrics_list
```

### <span style="color:blue">4.4 CROSS VALIDATION MODELLING</span> <a class="anchor" id="fourtysecond-bullet"></a>
K-fold cross validation may improve on the performance of training and test splits because it splits the entire dataset into K equal sections of data. At each iteration, it fits the model on different sections of the data and runs a test on the remaining test set, which must also be different for each iteration.

The scores are finally averaged out, which means below-average and above-average score may cancel each other out to give a more realistic score.

```python
# Create a Function for Cross Validation test
def cross_val_test(K_folds, model_instance, X, y):
    
    # Plot a distribution of train and test accuracies over a range of K folds
    train_accuracy = []
    test_accuracy = []

    for fold in K_folds:
        # Instantiate classifier model
        model = model_instance

        # Perform cross-validation on full dataset to evaluate model performance
        cv_scores = cross_validate(model, X, y, cv=fold, return_train_score=True)

        # Find mean train and test scores
        mean_train_score = np.mean(cv_scores['train_score'])
        mean_test_score = np.mean(cv_scores['test_score'])

        # Append the lists for mean scores
        train_accuracy.append(mean_train_score)
        test_accuracy.append(mean_test_score)

    # Plot the mean train and test scores against K fold
    plt.plot(K_folds, train_accuracy, label='Train score')
    plt.plot(K_folds, test_accuracy, label='Test score')
    plt.xlabel('Cross Validation Fold')
    plt.ylabel('Accuracy Score')
    plt.legend();
    return f"This is the Cross Validation Test for {model_instance}"
```

#### <span style="color:blue">4.4.1 LOGISTIC REGRESSION WITH CROSS VALIDATION</span> <a class="anchor" id="fourtythird-bullet"></a>

```python
# Plot a distribution of train and test accuracies over a range of K folds
cross_val_test(range(5, 18), LogisticRegression(random_state=42), X, y)
```

From the distribution of accuracy across different K folds, we observe the test accuracy score closely matches that of the train score. Therefore, the model is not overfitting.

### <span style="color:blue">4.7 BAGGED TREES CLASSIFIER</span> <a class="anchor" id="fiftyfirst-bullet"></a>

Bootstrap Aggregation does sampling with replacement to ensure that each tree in an ensemble fits on a different sample of the dataset compared to the other trees.

This will lead to normally distributed scores whose average may more closely approach the ground truth compared to a single decision tree classifier. 

I will fit a bagged classifier at the tuned maximum depth of 15 and 20 estimators.

```python
# Define a pipeline for Random Forest classification
pipe_bagged = Pipeline([('ss', StandardScaler()), 
                    ('bag', BaggingClassifier(DecisionTreeClassifier(criterion='gini', 
                                                    max_depth=15, min_samples_split=8, 
                                                    max_features='sqrt', random_state=42), 
                               n_estimators=20))])

# Fit to the training data
pipe_bagged.fit(X_train, y_train)

y_train_pred = pipe_bagged.predict(X_train)
y_test_pred = pipe_bagged.predict(X_test)

# Print accuracy and F1-scores
print("Bagged Tree Train Accuracy", pipe_bagged.score(X_train, y_train))
print("Bagged Tree Train F1-score", f1_score(y_train, y_train_pred, average='weighted'))

print("Bagged Tree Test Accuracy", pipe_bagged.score(X_test, y_test))
print("Bagged Tree Test F1-score", f1_score(y_test, y_test_pred, average='weighted'))
```

### <span style="color:blue">4.11 ROC-AUC ANALYSIS</span> <a class="anchor" id="sixtyfourth-bullet"></a>

```python
# Create a LabelBinarizer instance with multioutput='uniform'
lb = LabelBinarizer()

# Fit the label binarizer on the multiclass labels
lb.fit(y_test)

# Transform the multiclass labels into a binary format
y_multi_test = lb.transform(y_test)

# Predict probability for each model
log_model_pred_proba = pipe_log_new.predict_proba(X_test)
dt_model_pred_proba = pipe_decision_tree.predict_proba(X_test)
knn_model_pred_proba = pipe_knn_new.predict_proba(X_test)
bagged_model_pred_proba = pipe_bagged.predict_proba(X_test)
rf_model_pred_proba = pipe_rf.predict_proba(X_test)
xgb_model_pred_proba = pipe_xgb.predict_proba(X_test)

# Initialize variables to store AUC scores
log_model_auc = 0
dt_auc = 0
knn_auc = 0
bagged_auc = 0
rf_auc = 0
xgb_auc = 0

# Calculate AUC ROC score using predicted probability for each model
for class_idx in range(y_multi_test.shape[1]):  # Iterate over classes
    log_model_auc += roc_auc_score(y_multi_test[:, class_idx], log_model_pred_proba[:, class_idx])
    dt_auc += roc_auc_score(y_multi_test[:, class_idx], dt_model_pred_proba[:, class_idx])
    knn_auc += roc_auc_score(y_multi_test[:, class_idx], knn_model_pred_proba[:, class_idx])
    bagged_auc += roc_auc_score(y_multi_test[:, class_idx], bagged_model_pred_proba[:, class_idx])
    rf_auc += roc_auc_score(y_multi_test[:, class_idx], rf_model_pred_proba[:, class_idx])
    xgb_auc += roc_auc_score(y_multi_test[:, class_idx], xgb_model_pred_proba[:, class_idx])

# Calculate the mean AUC score over classes
num_classes = y_multi_test.shape[1]
log_model_auc /= num_classes
dt_auc /= num_classes
knn_auc /= num_classes
bagged_auc /= num_classes
rf_auc /= num_classes
xgb_auc /= num_classes

print(f'The AUC score for tuned Logistic Regression model is: {round(log_model_auc, 4)}')
print(f'The AUC score for baseline Decision Tree model is: {round(dt_auc, 4)}')
print(f'The AUC score for tuned K-Nearest-Neighbor model is: {round(knn_auc, 4)}')   
print(f'The AUC score for Bagged Tree model is: {round(bagged_auc, 4)}')
print(f'The AUC score for Random Forest model is: {round(rf_auc, 4)}')
print(f'The AUC score for XG Boost model is: {round(xgb_auc, 4)}')
```

```python
from sklearn.metrics import roc_curve, auc

# List of models
models = [pipe_log_new, pipe_decision_tree, pipe_knn_new, 
          pipe_bagged, pipe_rf, pipe_xgb]
model_names = ['Logistic Regression', 'Decision Tree', 
               'K-Nearest Neighbor', 'Bagged Tree', 'Random Forest', 'XG Boost']

# Specify the number of target classes
n_class = 3

# Initialize the subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.ravel()

for i, model in enumerate(models):
    # Create a dictionary for colors and labels
    class_info = {
        0: {'color': 'red', 'label': 'Non Functional'},
        1: {'color': 'green', 'label': 'Functional Needs Repair'},
        2: {'color': 'blue', 'label': 'Functional'}
    }
    
    # Iterate through each class to create roc_curve
    for j in range(n_class):
        fpr_i, tpr_i, _ = roc_curve(y_multi_test[:, j], 
                                    model.predict_proba(X_test)[:, j])
        roc_auc_i = auc(fpr_i, tpr_i)

        # Plot the ROC curve for the current class with a specific color and label
        axes[i].plot(fpr_i, tpr_i, color=class_info[j]['color'], 
                     lw=2, label=f'{class_info[j]["label"]} (AUC = {roc_auc_i:.2f})')
    
    # Title and axis labels
    axes[i].set_title(f'{model_names[i]} ROC curve')
    axes[i].set_xlabel('False Positive Rate')
    axes[i].set_ylabel('True Positive Rate')
    axes[i].legend(loc='best')

# Adjust the layout
plt.tight_layout()
plt.show()
```

## <span style="color:blue">STUDY LIMITATIONS</span> <a class="anchor" id="sixtyeighth-bullet"></a>

1. About half of the dataset has zero population, gps height, total static head, and construction year values and there is no information about why this is the case. This has only resulted in unresolved dilemmas. For example, wells with zero population have the highest frequency of non-functional pumps, but there is no information about whether the wells have been abandoned or the population has relocated. Therefore, I cannot accurately answer questions such as the relationship between population or construction year and a pump's functionality.

2. There is a high number of pumps with zero static head, the majority of which are non-functional. It is not possible to conclude whether the pumps are actually faulty or they are in good condition but the wells have run dry. 

3. Tuning the Machine Learning models was computationally intensive and some of them took a couple of hours to execute. Therefore, I could not tune combinations of all possible parameters in due time. 

4. The target classes were highly imbalanced and oversampling the minority class did not improve prediction scores.
